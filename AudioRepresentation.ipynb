```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display
```

```python
guitarTune, srG = librosa.load('guitarTune.wav')
babyCry, srB = librosa.load('babyCry.mp3')
```

```python
import IPython.display as ipd
print("Guitar Tune")
ipd.Audio(guitarTune, rate=srG)
```

```python
print("Baby Crying")
ipd.Audio(babyCry, rate=srB)
```

```python
plt.subplot(2, 1, 1)
librosa.display.waveshow(guitarTune, sr=srG)
plt.title("Guitar Tune in time domain")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")

plt.subplot(2, 1, 2)
librosa.display.waveshow(babyCry, sr=srB)
plt.title("Baby Crying sound in time domain")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.tight_layout()
plt.show()
```

```python
from scipy.fft import fft, ifft
plt.subplot(2, 1, 1)
N = len(guitarTune)
f = np.linspace(0.0, 1.0*srG/2, N//2)
plt.plot(f, abs(fft(guitarTune/N))[:N//2])
plt.title("Guitar Tune in frequency domain")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")

plt.subplot(2, 1, 2)
N = len(babyCry)
f = np.linspace(0.0, 1.0*srG/2, N//2)
plt.plot(f, abs(fft(babyCry)/N)[:N//2])
plt.title("Baby Crying sound in frequency domain")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.tight_layout()
plt.show()
```

```python
from scipy.fft import fft, ifft
plt.subplot(2, 1, 1)
spectrogram = librosa.amplitude_to_db(np.abs(librosa.stft(guitarTune)))
librosa.display.specshow(spectrogram, y_axis="log")
plt.title("Time-frequency representation of guitar tune")
plt.xlabel("Time")
plt.ylabel("Frequency (Hz) (Log scale)")

plt.subplot(2, 1, 2)
spectrogram = librosa.amplitude_to_db(np.abs(librosa.stft(babyCry)))
librosa.display.specshow(spectrogram, y_axis="log")
plt.title("Time-frequency representation of baby crying sound")
plt.xlabel("Time")
plt.ylabel("Frequency (Hz) (Log scale)")
plt.tight_layout()
plt.show()
```
